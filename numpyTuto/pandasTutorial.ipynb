```python
import pandas as pd

#How to make series

data = [100, 102, 101, 104]

series = pd.Series(data, index = ["a", "b", "c", "d"])

series.loc["c"] = 200


print(series.loc["c"])
print(series)
#will return the value of whatever is in index a

print(series.iloc[2])
#this will print data type in the index
#similar to an array
#by integer location

filData = [100, 110, 103, 101, 114]

series2 = pd.Series(filData, index = ["a", "b", "c", "d", "e"])
print(series2[series2 > 109])
```

```python
calories = {"Day 1": 1750, "Day 2": 2100, "Day 3": 1700}

series = pd.Series(calories)

series.loc["Day 3"] += 500

print(series.loc["Day 3"])
print(series[series < 2000])
```

```python
#My example

footballers = {"Messi": True, "Ronaldo": True, "Zidane": True, "Zlatan": False, "Vini": False}

series = pd.Series(footballers)

print("The following players have won the Ballon D'or!!!!")
print(series[series == True])
```

```python
#Data Frames is two dimensional

data = {
        "Name": ["SpongeBob", "Patrick", "Squidward"],
        "Age": [30, 35, 50]
        }

df = pd.DataFrame(data, index = ["Employee 1", "Employee 2", "Employee 3"])
print(df)

print('')
print(df.loc["Employee 2"])

print('')
print(df.iloc[0])

# add a new column
df["Job"] = ["Cook", "N/A", "Cashier"]
print(df)

#add a new rows
new_rows = pd.DataFrame([{"Name": "Sandy", "Age": 28, "Job": "Engineer"},
                        {"Name": "Mr.Crabs", "Age": 58, "Job": "Manager"}], 
                       index =["Employee 4", "Employee 5"])
df = pd.concat([df, new_rows])
print(" ")
print(df)
```
